2. Пожалуйста, реализуйте функцию ```split_data(data, ratio=0.1)```, которая разбивает набор данных, загруженный в предыдущем вопросе, на наборы данных ```train``` и ```validation``` в соотношении 9:1.

````python
import numpy as np

def split_data(data, ratio=0.1):
    np.random.seed(1)
    #Начало вашего кода    
    indeces = np.arange(0,len(data))
    np.random.shuffle(indeces)
    val_ind = indeces[: round(ratio * len(indeces))]
    train_ind = indeces[round(ratio * len(indeces)):]
    val_data = data.loc[val_ind]
    train_data = data.loc[train_ind]
    #Конец вашего кода
    return {"validation":val_data, "train":train_data}

splitted_data = split_data(num_data, ratio=0.1)
````

5. Теперь пришло время использовать случайный лес (регрессии) для параметров в списке ```important``` для прогнозирования ```SalePrice```. Пожалуйста, реализуйте функцию ```train_random_forest_regressor(data, parameters, target="SalePrice", n_estimators=1, random_state=0)```, которая будет обучать случайный лес (регрессии) для параметров из списка ```parameters```. Функция должна возвращать словарь с обученным случайным лесом (регрессии), списком ```parameters```.

````python
#Начало вашего кода 
def train_random_forest_regressor(data, parameters, target="SalePrice", n_estimators=1, random_state=0):
    from sklearn.ensemble import RandomForestRegressor
    regressor = RandomForestRegressor(n_estimators = n_estimators, random_state = random_state)
    regressor.fit(data[parameters].values,  data[target])
    d = {}
    d['regressor'] = regressor
    d['parameters'] = parameters
    return d
#Конец вашего кода
model = train_random_forest_regressor(splitted_data["train"], important, target="SalePrice", n_estimators=1)
print(model)
````

6. Пожалуйста, реализуйте функцию ```validation(model, data, target="SalePrice")```, которая будет принимать словарь ```model```, данные ```validation``` и параметр ```target```, который должен быть предсказан. Функция должна возвращать ```DataFrame``` с первым столбцом, который будет истинным ```"SalePrice"```,  а второй столбец предсказан случайным лесом и ```rmse```.

````python
#Начало вашего кода
def validation(model, data, target="SalePrice"):
    pred = model['regressor'].predict(data[model['parameters']].values)
    df = pd.DataFrame()
    df['True Sale Price'] = data['SalePrice']
    df['Predicted Sale Price'] = pred
    rmse = rmse = (np.mean(( data['SalePrice']- pred)**2)) **0.5
    return df, rmse
#Конец вашего кода
````

**Примечание:** figsize=(8,8)

````python
#Начало вашего кода
df , rmse = validation(model,splitted_data["train"], target="SalePrice" )
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (8,8))
sns.set_context('notebook');
sns.scatterplot(x = df['True Sale Price'], y = df['Predicted Sale Price'] );
plt.title(f'Train dataset rmse {round(rmse, 2)} USD');
#Конец вашего кода
````

````python
#Начало вашего кода
# Дерево принятия решений 
def train_decision_tree_regressor(data, parameters, target="SalePrice", degree=1, random_state=0):
    from sklearn.tree import DecisionTreeRegressor 
    regressor = DecisionTreeRegressor(random_state = random_state)
    regressor.fit(data[parameters].values, data[target])
    d = {}
    d['regressor'] = regressor
    d['parameters'] = parameters
    return d


def validation(model, data, target="SalePrice"):
    df = pd.DataFrame()
    pred = model['regressor'].predict(data[model['parameters']].values)
    df['True Sale Price'] = data['SalePrice']
    df['Predicted Sale Price'] = pred
    rmse = rmse = (np.mean(( data['SalePrice']- pred)**2)) **0.5
    return df, rmse
model_tree = train_decision_tree_regressor(splitted_data["train"], important, target="SalePrice")
df , rmse = validation(model_tree,splitted_data["train"], target="SalePrice" )
print(rmse)
#Конец вашего кода
````
