5. Пожалуйста, реализуйте функцию ```plot_f_df(f,df,xlims=(-10,10))```, которая будет отображать функцию ```f``` и производную ```df``` в диапазоне xlims по оси x.

````python
#Начало вашего кода
import seaborn as sns
import matplotlib.pyplot as plt
def plot_f_df(f,df,xlims=(-10,10)):
    x = np.arange(xlims[0], xlims[1]+1, 0.5)
    y = f(x)
    plt.figure(figsize = (10,5))
    plt.subplot(1,2,1)
    sns.lineplot(x, y, label = 'f')
    sns.set_context('talk')
    sns.set_style('whitegrid')
    plt.xlim(xlims)
    y = df(x)

    
    plt.subplot(1,2,2)
    plt.plot(x, y, 'r:', label= 'df')
    plt.legend()
    plt.xlim(xlims)
#Конец вашего кода

plot_f_df(lambda x: -x**2+10*x, lambda x: -2*x+10)
````
