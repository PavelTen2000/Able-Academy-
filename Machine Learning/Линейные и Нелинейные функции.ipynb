    * –ö–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä—É–µ—Ç –Ω–∞–∫–ª–æ–Ω —Ñ—É–Ω–∫—Ü–∏–∏? –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```ans3```.
    
    * –ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —Å —Ñ—É–Ω–∫—Ü–∏–µ–π, –µ—Å–ª–∏ ùíé=ùüé? –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```ans4```.

    * $y=x+1/x-20$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y1```.
    
    * $y=x^2/x-5x$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y2```.
    
    * $y=1/log(x)$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y3```.

    * $y=(x-4)^2$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y5```.

    * $y=1-x$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y8```.

3. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é ```rate_of_change(x1,x2,y1,y2)```, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —Å–∫–æ—Ä–æ—Å—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è (rate of change = RoC) –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å 4 –ø–∞—Ä–∞–º–µ—Ç—Ä–∞, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö 2 x-–∑–Ω–∞—á–µ–Ω–∏—è –∏ 2 y-–∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è 2 —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–µ–∫: ```x1, x2, y1, y2```:

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
def rate_of_change(x1,x2,y1,y2):
    return (y2-y1)/ ( x2 -x1)
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

print(f"RoC: {rate_of_change(0,1,0,1)}")
print(f"RoC: {rate_of_change(0,1,1,0)}")
print(f"RoC: {rate_of_change(10,3,-10,4)}")
````

4. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é ```plot_secant(fun,x1,x2,xlims=(-10,10),ylims=(-100,200))```, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é ```fun``` –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ xlims –ø–æ –æ—Å–∏ x –∏ ylims –ø–æ –æ—Å–∏ y. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, —Å–ª–µ–¥—É–µ—Ç –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å–µ–∫—É—â—É—é –ª–∏–Ω–∏—é (secant), —Å–æ–µ–¥–∏–Ω—è—é—â—É—é —Ç–æ—á–∫–∏ ```x1``` –∏ ```x2``` (–∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏, –∫–∞–∫ –≤ –≤–∏–¥–µ–æ–ª–µ–∫—Ü–∏–∏).

    **–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** –≤–∞–º –º–æ–∂–µ—Ç –ø–æ–Ω–∞–¥–æ–±–∏—Ç—å—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é ```rate_of_change()```.

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def plot_secant(fun,x1,x2,xlims=(-10,10),ylims=(-100,200)):
    x = np.arange(xlims[0] ,xlims[1]+1)
    y = fun(x)
    sns.set_context('talk')
    sns.set_style('whitegrid')
    sns.lineplot(x = x, y = y, label = 'function')
    plt.xlim(xlims)
    plt.ylim(ylims)
    m = rate_of_change(x1,x2, fun(x1),fun(x2))
    b = fun(x1) - m*x1
    sec_fun = lambda x: m*x + b 
    sns.lineplot(x = x, y = sec_fun(x), label = f'y = {round(m, 1)}x + {round(b, 1)}')
    plt.title(f'x1 = {x1} & x2 = {x2}')
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

y = lambda x: -x**2+2*x+20
plot_secant(y,-5,3)
````

5. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –≤–æ—Å–ø—Ä–æ–∏–∑–≤–µ—Å—Ç–∏ –≥—Ä–∞—Ñ–∏–∫–∏, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–µ –≤ –∫–æ–Ω—Ü–µ –≤–∏–¥–µ–æ–ª–µ–∫—Ü–∏–∏, –Ω–æ —Å –¥—Ä—É–≥–æ–π –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–π –Ω–∏–∂–µ.
    
    **–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** –≤–∞–º –º–æ–∂–µ—Ç –ø–æ–Ω–∞–¥–æ–±–∏—Ç—å—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é ```plot_secant()```. –†–∞–∑–º–µ—Ä —Ñ–∏–≥—É—Ä—ã —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –Ω–∞ ```(15,15)```.
    
### –û–∂–∏–¥–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:



````python
y = lambda x: -x**2+10
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
xes = [(-5,5) , (-5,3) , (-5,1), (-5,0), (-5,-1), (-5,-2), (-5,-3) , (-5,-4), (-5, -4.9)]
plt.figure(figsize = (15,15))
for num, val in enumerate(xes, 1):
    plt.subplot(3,3,num)
    plot_secant(y,val[0], val[1])
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
````
