3. Пожалуйста, смоделируйте эксперименты с монетой. Предположим, что есть одинаковый шанс получить Heads и Tails после одного броска монеты. Оцените вероятность получения Tails, если вы подбрасываете монету от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:




````python
#Начало вашего кода
x = np.arange(10,10001)
y = np.array([])
np.random.seed(1)
for i in x:
    a = np.random.randint(2, size = (i))
    y = np.append(y, (a.sum()/ i ) )
plt.figure(figsize = (7,5))
plt.plot(x, y , linestyle = ':', label = 'Simulation')
plt.yticks(np.arange(0,1.1,0.1));
plt.xticks([10,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]);
plt.title('P(tails)')
plt.xlim(0,10000)
plt.axhline(y = 0.5 , ls = ':', label = 'Theoretical', c = 'r')
plt.ylabel('Probability of Tails')
plt.xlabel('# of flips')
plt.legend();
#Конец вашего кода
````

4. Пожалуйста, смоделируйте эксперименты с игральной костью (со сторонами ```{1,2,3,4,5,6}```). Предположим, что есть равный шанс получить любую из шести сторон после броска игральной кости. Оцените вероятность получить число меньше или равное 2 после броска игральной кости от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:



````python
#Начало вашего кода
x = np.arange(10,10001)
y = np.array([])
np.random.seed(1)
for i in x: 
    a = np.random.randint(1,7, size = i)
    lessthanthree = (a < 3).sum() / i
    y = np.append(y, lessthanthree)
plt.figure(figsize = (7,5))
plt.plot(x, y , linestyle = ':', label = 'Simulation')
plt.yticks(np.arange(0,1.1,0.1));
plt.xticks([10,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]);
plt.title('P(<3)')
plt.xlim(0,10000)
plt.axhline(y = 2/6 , ls = ':', label = 'Theoretical', c = 'r')
plt.ylabel('Probability')
plt.xlabel('# of flips')
plt.legend();
#Конец вашего кода
````
