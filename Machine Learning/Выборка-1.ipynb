Пожалуйста, реализуйте функцию ```random_samples(data, sample_size=20, column=None)```, которая будет делать случайную выборку из данных ```data``` размера ```sample_size``` и вычислять статистику для ```column```. Функция должна возвращать список seeds и список статистики соответствующих выборок.

````python
#Начало вашего кода
def random_samples(data, sample_size=20, column=None):
    seeds = []
    means = []
    for i in range(1,100):
        np.random.seed(i)
        sample = data.iloc[np.random.randint(0, data[column].size, sample_size )    , :]
        mean = sample[column].mean()
        seeds.append(i)
        means.append(mean)
    return seeds, means
#Конец вашего кода

seeds, means = random_samples(data,sample_size=5, column="age")

print(seeds)
print(means)
````

8. Используя функцию ```random_samples```, визуализируйте ошибку выборки (```sample_size = 20```) для различных ```random seeds``` столбцов ```"age"```, ```"player_height"``` и ```"player_weight"```.

````python
#Начало вашего кода
plt.figure(figsize = (15,5))
for num, val in enumerate(["age", "player_height", "player_weight"], 1):
    plt.subplot(1,3, num)
    seeds , means = random_samples(data,sample_size=20, column=val)
    error = np.array(means) - data[val].mean()
    sns.scatterplot(x =seeds , y = error )
    if val == 'age':
        plt.title(f'{val} sampling error')
        plt.ylabel(val)
    else:
        plt.title(f'{val.strip("player_")} sampling error')
        plt.ylabel(val.strip("player_"))
    plt.xlabel('random seed')
#Конец вашего кода
````
