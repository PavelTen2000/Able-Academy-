1. –í—ã —Ä–∞–±–æ—Ç–∞–µ—Ç–µ –≤ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º —Ü–µ–Ω—Ç—Ä–µ. –ë—ã–ª —Ä–∞–∑—Ä–∞–±–æ—Ç–∞–Ω –Ω–æ–≤—ã–π —Ç–µ—Å—Ç –Ω–∞ $HIV$. –í–∞—Å –ø—Ä–æ—Å—è—Ç –æ—Ü–µ–Ω–∏—Ç—å –Ω–æ–≤—ã–π —Ç–µ—Å—Ç, —Ö–æ—Ä–æ—à–æ –ª–∏ –æ–Ω —Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤–∞—à–∏—Ö –∑–Ω–∞–Ω–∏–π –ø–æ —Ç–µ–æ—Ä–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π. –í–∞–º –¥–∞–Ω—ã —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –ø–æ –Ω–æ–≤–æ–º—É —Ç–µ—Å—Ç—É –Ω–∞ $HIV$.
    
    –£—Å–ª–æ–≤–Ω—ã–µ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:
    
    * $HIV^+$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π, –∏–Ω—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö $HIV$.
    
    * $HIV^-$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π, –ù–ï –∏–Ω—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö $HIV$.
    
    * $Test^+$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π, —É –∫–æ—Ç–æ—Ä—ã—Ö —Ç–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ $HIV$.
    
    * $Test^-$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π, —É –∫–æ—Ç–æ—Ä—ã—Ö —Ç–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ $HIV$.


||$HIV^+$|$HIV^-$|Total|
|:-:|:-:|:-:|:-:|
|$Test^+$|10 |9    |19   |
|$Test^-$|7  |123  |130  |
|Total   |17 |132  |149  |


–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –≤–∞–º –¥–∞–Ω–Ω—ã—Ö, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏::

* $P(Test^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tp```.
* $P(Test^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tn```.
* $P(HIV^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_hp```.
* $P(HIV^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_hn```.
* $P(Test^+|HIV^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tp_hp```.
* $P(Test^-|HIV^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tn_hp```.
* $P(Test^+|HIV^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tp_hn```.
* $P(Test^-|HIV^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_tn_hn```.

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
p_tp = 19 / 149
p_tn = 130 / 149
p_hp = 17 / 149
p_hn = 132 / 149
p_tp_hp = 10/17
p_tn_hp = 7 / 17
p_tp_hn = 9 / 132
p_tn_hn = 123 / 132
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

print(f"{p_tp*100:.2f}%")
print(f"{p_tn*100:.2f}%")
print(f"{p_hp*100:.2f}%")
print(f"{p_hn*100:.2f}%")
print(f"{p_tp_hp*100:.2f}%")
print(f"{p_tn_hp*100:.2f}%")
print(f"{p_tp_hn*100:.2f}%")
print(f"{p_tn_hn*100:.2f}%")
````

2. –≠—Ç–æ—Ç –≤–æ–ø—Ä–æ—Å —Å–≤—è–∑–∞–Ω —Å –≤–æ–ø—Ä–æ—Å–æ–º 1. 
    –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–∞–Ω–Ω—ã—Ö, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã—Ö –≤ —Ç–∞–±–ª–∏—Ü–µ –≤—ã—à–µ, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:

    * $P_{table}(HIV^+|Test^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_t_hp_tp```.
    * $P_{table}(HIV^+|Test^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_t_hp_np```.

    –û—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–µ–æ—Ä–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –≤–∞—à–µ–≥–æ –∑–Ω–∞–Ω–∏—è –∏ –∏—Å–ø–æ–ª—å–∑—É—è –∏–∑–≤–µ—Å—Ç–Ω—ã–µ –≤–∞–º —Ñ–æ—Ä–º—É–ª—ã, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:

    * $P_{formula}(HIV^+|Test^+)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_f_hp_tp```.
    * $P_{formula}(HIV^+|Test^-)$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_f_hp_np```.


````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
p_t_hp_tp = 10 / 19
p_t_hp_tn = 7 / 130
p_f_hp_tp = (p_hp * p_tp_hp) / ((p_hp *  p_tp_hp) + (p_hn * p_tp_hn))
p_f_hp_tn = (p_hp * p_tn_hp) / ((p_hp * p_tn_hp) + (p_hn * p_tn_hn))
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
print("–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–∞–Ω–Ω—ã—Ö:")
print(f"{p_t_hp_tp*100:.2f}%")
print(f"{p_t_hp_tn*100:.2f}%")

print("–ò—Å–ø–æ–ª—å–∑—É—è —Ñ–æ—Ä–º—É–ª—ã:")
print(f"{p_f_hp_tp*100:.2f}%")
print(f"{p_f_hp_tn*100:.2f}%")
````

* $P("urgent")$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_urgent```.
* $P("spam"|"urgent")$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_spam_given_urgent```.
* $P("non-spam"|"urgent")$, –∏ –ø—Ä–∏—Å–≤–æ–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```p_non_spam_given_urgent```.


````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
p_spam = 0.27
p_non_spam = 1 - p_spam
p_urgent_given_spam = 0.46
p_urgent_given_non_spam = 0.19
p_urgent_and_spam = p_spam * p_urgent_given_spam
p_urgent_and_non_spam = p_non_spam * p_urgent_given_non_spam
p_urgent = p_spam * p_urgent_given_spam + p_non_spam * p_urgent_given_non_spam
p_spam_given_urgent = p_urgent_and_spam / p_urgent
p_non_spam_given_urgent = p_urgent_and_non_spam / p_urgent
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

print(f"{p_spam*100:.2f}%")
print(f"{p_non_spam*100:.2f}%")
print(f"{p_urgent_given_spam*100:.2f}%")
print(f"{p_urgent_given_non_spam*100:.2f}%")
print(f"{p_urgent_and_spam*100:.2f}%")
print(f"{p_urgent_and_non_spam*100:.2f}%")
print(f"{p_urgent*100:.2f}%")
print(f"{p_spam_given_urgent*100:.2f}%")
print(f"{p_non_spam_given_urgent*100:.2f}%")
````

5. –í—ã –≤—ã—á–∏—Å–ª–∏–ª–∏ —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º –≤–æ–ø—Ä–æ—Å–µ: $P("spam")$, $P("spam"|"urgent")$.

    * –ü—Ä–∏—Å–≤–æ–π—Ç–µ –∞–ø—Ä–∏–æ—Ä–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∞–º –ø–æ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π: ```prior```.
    
    * –ü—Ä–∏—Å–≤–æ–π—Ç–µ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∞–º –ø–æ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π: ```posterior```.
    
    * –†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –º–µ–∂–¥—É –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–π –∏ –∞–ø—Ä–∏–æ—Ä–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é - –≤–∞–º –Ω—É–∂–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–∞ –∞–ø—Ä–∏–æ—Ä–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å. –ü—Ä–∏—Å–≤–æ–π—Ç–µ –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π: ```ratio```.

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
prior = p_spam
posterior = p_spam_given_urgent
ratio = posterior / prior
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
print(f"{prior*100:.2f}%")
print(f"{posterior*100:.2f}%")
print(f"{ratio:.2f}")
````
