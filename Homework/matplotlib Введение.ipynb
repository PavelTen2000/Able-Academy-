### Шаг 5

Пожалуйста, создайте ```pie``` с помощью ```rand_nums``` (созданную ранее).

**Подсказка: проверьте входной аргумент ```explode```.** 
Должен быть установлен на ```0.1```.

````python
#Начало вашего кода
unique , count = np.unique(rand_nums, return_counts=True)
explode = np.repeat(0.1,10)
plt.pie(count , labels = unique, autopct = lambda x: f'{x:.1f}%', explode = explode);
plt.title('Pie chart');

#Конец вашего кода
````
