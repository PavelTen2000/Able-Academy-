### Шаг 4

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода
sns.set_context('talk')
h = sns.FacetGrid(titanic, col = 'alone', row = 'sex', hue = 'alive', height = 5, hue_order = ['yes', 'no']);
h.map(sns.scatterplot, 'age', 'fare').add_legend();
#Конец вашего кода
````

### Шаг 5

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода
sns.set(style = 'darkgrid')
sns.set_context('talk')
h = sns.FacetGrid(titanic, col = 'pclass', row = 'alone', hue = 'alive', height = 5, hue_order = ['yes', 'no'],
                 row_order = [True, False]);
h.map(sns.swarmplot, 'sex', 'age').add_legend();
#Конец вашего кода
````
