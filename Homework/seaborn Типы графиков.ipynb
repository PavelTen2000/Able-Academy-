### Шаг 5

Пожалуйста, реализуйте функцию ```plot_sns_age_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_age_pclass(data=None):
    #Начало вашего кода
    for i in set(data['pclass']):
        sns.catplot(x = 'sex', y = 'age', hue = 'alive', hue_order = ['yes', 'no']  , data = data[data['pclass'] == i],
                   height = 4, aspect = 8/4)
        plt.title(f'Passenger class: {i}')
    #Конец вашего кода

plot_sns_age_pclass(data=titanic)
````

### Шаг 6

Пожалуйста, реализуйте функцию ```plot_sns_fare_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_fare_pclass(data=None):
    #Начало вашего кода
    for i in set(data['pclass']):
        plt.rcParams['figure.figsize'] = 8,4
        sns.swarmplot(x= 'sex', y = 'fare', hue = 'alive', hue_order = ['yes','no'] ,data = data[data['pclass'] == i])
        plt.title(f'Passenger class: {i}')
        plt.show()
    #Конец вашего кода

plot_sns_fare_pclass(data=titanic)
````

### Шаг 7

Пожалуйста, реализуйте функцию plot_counts(data=None), которая отображает несколько фигур, каждая фигура содержит визуализацию на основе ```param```.

````python
def plot_counts(data=None,param=[],fontsize=15):
    #Начало вашего кода
    for i in param:
        sns.countplot(x = 'alive', hue = i, data = titanic)
        plt.title(f'Counts based on {i}', fontsize = fontsize)
        plt.show()
    #Конец вашего кода

plot_counts(data=titanic,param=["pclass","sex"])
````

### Шаг 8

Пожалуйста, реализуйте функцию ```plot_dists(data=None,param=[],fontsize=15)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе ```param```.


````python
def plot_dists(data=None,param=[],fontsize=15):
    #Начало вашего кода
    for i in param:
        sns.distplot(data[i])
        plt.title(f'{i} distribution', fontsize = fontsize)
        plt.show()
    #Конец вашего кода

plot_dists(data=titanic,param=["fare","age"])
````
